```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
```

```python
URL = 'https://realpython.github.io/fake-jobs/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64'}


response = requests.get(URL, headers = headers)
```

```python
soup = BS(response.text)
```

```python
print(soup.prettify())
```

```python
soup.find('h2').text
```

```python
titles = soup.findAll('h2')
```

```python
jobtitles = [title.text for title in titles]
jobtitles
```

```python
companies = soup.findAll('h3')
```

```python
jobcompanies = [company.text for company in companies]
jobcompanies
```

```python
locations = soup.findAll('p', attrs={'class' : 'location'})
```

```python
joblocation = [location.text.strip() for location in locations]
joblocation
```

```python
postingdate = soup.findAll('time', attrs={'datetime' : '2021-04-08'})
```

```python
date_clean = [date.text.strip() for date in postingdate]
```

```python
date_clean
```

```python
jobsdataframe = pd.DataFrame({'jobtitles': jobtitles, 'company': jobcompanies, 'location': joblocation, 'date_opened': date_clean})
```

```python
jobsdataframe
```

```python
#a class="card-footer-item"
buttonurl = soup.findAll('a')
```

```python
buttonurl
```

```python
apply_urls = [a['href'] for a in soup.findAll('a', string='Apply')]
```

```python
apply_urls
```

```python
jobsdataframe['applyurl'] = apply_urls
jobsdataframe
```

```python
jobURL = 'https://realpython.github.io/fake-jobs/jobs/senior-python-developer-0.html'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64'}


response2 = requests.get(jobURL, headers = headers)
```

```python
jobs = BS(response2.text)
```

```python
print(jobs.prettify())
```

```python
jobdescription = [div.find('p').text.strip() for div in jobs.findAll('div', attrs={'class': 'content'}) if div.find('p')]
```

```python
jobdescription
```

```python
def get_job_description(job_url):
    try:
        response = requests.get(job_url)
        response.raise_for_status()
        job_soup = BS(response.content, 'html.parser')
        desc_tag = job_soup.find('div', class_= 'content').find('p')
        return desc_tag.text.strip() if desc_tag else "No description found"
    except Exception as e:
        return f"Error: {e}"
```

```python
get_job_description('https://realpython.github.io/fake-jobs/jobs/television-floor-manager-8.html')
```

```python
jobsdataframe['description'] = jobsdataframe['applyurl'].apply(get_job_description)
```
